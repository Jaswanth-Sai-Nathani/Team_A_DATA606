## WEBSCRAPING THE CARS.COM FOR GETTING THE DATA OF USED CARS

### Importing required packages for scraping

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

### Accessing the cars.com website and walking through different pages in the website

```python
cars_pages_list = [] 
for idx, link in enumerate(range(10)):
    #url = "https://www.cars.com/shopping/results/?page=" + str(idx + 1) + "&page_size=20&list_price_max=&makes[]=&maximum_distance=10&models[]=&stock_type=used&zip=21227"
    url = 'https://www.cars.com/shopping/results/?page=' + str(idx + 1) + '&page_size=20&list_price_max=&makes[]=&maximum_distance=100&models[]=&stock_type=all&zip=21227'
    cars_pages_list.append(url)
print(len(cars_pages_list))
cars_pages_list
```

### Accessing link of each car for sale

```python
base_url = 'https://www.cars.com'
all_cars_urls = []
for link in cars_pages_list:
    req = requests.get(link)

    soup = BeautifulSoup(req.text,'html.parser')

    req_links = soup.find_all('a')


    for link in req_links:
        suffix = str(link.get('href'))
        if ('/vehicledetail/' in suffix):
            car_url = base_url + suffix
            all_cars_urls.append(car_url)
```

### Formatting the obtained links to avoid duplicates

```python
set_all_cars_urls = set(all_cars_urls)
final_urls = list(set_all_cars_urls)
len(final_urls)
```

### Adopting the columns for creating a dataframe

```python
make=[]
model=[]
trim=[]
year = []
stock_type = []
mileage = []
dealer_name = []
distance = []
rating = []
review_count = []
badge_label = []
price = []

i=0
for output in output_details:
    

    current_car = output.find('h2').get_text()
    
    try:
        make.append(current_car.split(' ')[1])
    except:
        make.append('n/a')
    
    try:
        model.append(current_car.split(' ')[2])
    except:
        model.append('n/a')
    
    try:
        edition = current_car.split(' ')[3:]
        nt=''
        for var in edition:
            nt += ' ' + var
        trim.append(nt)
    except:
        trim.append('n/a')
    
    try:
        year.append(current_car.split(' ')[0])
    except:
        year.append('n/a')
        
    try:
        stock_type.append(output.find('p').get_text())
    except:
        stock_type.append('n/a')
    
    # mileage
    try:
        mileage.append(output.find('div', {'class':'mileage'}).get_text())
    except:
        mileage.append('n/a')
    
    # dealer_name
    try:
        dealer_name.append(output.find('div', {'class':'dealer-name'}).get_text().strip())
    except:
        dealer_name.append('n/a')
        
    try:
        distance.append(output.find('div',{'class':'miles-from'}).get_text().strip())
    except:
        distance.append('n/a')    
        
    # rating
    try:
        rating.append(output.find('span', {'class':'sds-rating__count'}).get_text())
    except:
        rating.append('n/a')
    
    # review_count
    try:
        review_count.append(output.find('span', {'class':'sds-rating__link'}).get_text())
    except:
        review_count.append('n/a')
        
    try:
        badge_label.append(output.find('span',{'class':'sds-badge__label'}).get_text())
    except:
        badge_label.append('n/a')
    
    #price 
    try:
        price.append(output.find('span', {'class':'primary-price'}).get_text())
    except:
        price.append('n/a')
        
```

### Obtaining the values of the dataset with results of one page list

```python

Used_Cars = pd.DataFrame({'Brand': make, 'Model': model, 'Trim' : trim, 'Manufature Year': year, 
                           'Stock Type' : stock_type, 'Mileage':mileage, 'Dealer Name':dealer_name,
                           'Distance' : distance, 'Rating': rating, 'Review Count': review_count,
                           'Badge Label' : badge_label,'Price': price})
```

### Following the same method to obtain all the values of the details of cars on the website

```python
make=[]
model=[]
trim=[]
year = []
stock_type = []
mileage = []
dealer_name = []
distance = []
rating = []
review_count = []
badge_label = []
price = []

filename = 'Used Cars Prices'

for i in range (1,550):
    
    
    website = 'https://www.cars.com/shopping/results/?page=' +str(i) + '&page_size=20&list_price_max=&makes[]=&maximum_distance=10&models[]=&stock_type=used&zip=21227'
    
    
    response = requests.get(website)
   
    soup = BeautifulSoup(response.content, 'html.parser')
    
    output_details = soup.find_all('div', {'class' : 'vehicle-card'})
    
    
    for output in output_details:
        current_car = output.find('h2').get_text()
    
        try:
            make.append(current_car.split(' ')[1])
        except:
            make.append('n/a')
      
        try:
            model.append(current_car.split(' ')[2])
        except:
            model.append('n/a')
    
        try:
            edition = current_car.split(' ')[3:]
            nt=''
            for var in edition:
                nt += ' ' + var
            trim.append(nt)
        except:
            trim.append('n/a')
    
        try:
            year.append(current_car.split(' ')[0])
        except:
            year.append('n/a')
        
        try:
            stock_type.append(output.find('p').get_text())
        except:
            stock_type.append('n/a')
            
        try:
            mileage.append(output.find('div', {'class':'mileage'}).get_text())
        except:
            mileage.append('n/a')


        try:
            dealer_name.append(output.find('div', {'class':'dealer-name'}).get_text().strip())
        except:
            dealer_name.append('n/a')

        try:
            distance.append(output.find('div',{'class':'miles-from'}).get_text().strip())
        except:
            distance.append('n/a')
            

        try:
            rating.append(output.find('span', {'class':'sds-rating__count'}).get_text())
        except:
            rating.append('n/a')

            
        try:
            review_count.append(output.find('span', {'class':'sds-rating__link'}).get_text())
        except:
            review_count.append('n/a')

        try:
            badge_label.append(output.find('span',{'class':'sds-badge__label'}).get_text())
        except:
            badge_label.append('n/a')
            

        try:
            price.append(output.find('span', {'class':'primary-price'}).get_text())
        except:
            price.append('n/a')
            
            
```

```python
Used_Cars = pd.DataFrame({'Brand': make, 'Model': model, 'Trim' : trim, 'Manufature Year': year, 
                           'Stock Type' : stock_type, 'Mileage':mileage, 'Dealer Name':dealer_name,
                           'Distance' : distance, 'Rating': rating, 'Review Count': review_count,
                           'Badge Label' : badge_label,'Price': price})
```
