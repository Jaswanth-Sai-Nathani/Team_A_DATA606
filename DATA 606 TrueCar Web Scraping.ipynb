### Importing required packages for scraping

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import lxml.html as lxl
import numpy as np
```

### Accessing the cars.com website and walking through different pages in the website

```python
pages=[]
base_url = 'https://www.truecar.com/used-cars-for-sale/listings/'
for i in range(100):
        pages.append(base_url + '?page=' + str(i))
```

```python
pages
```

### Accessing link of each car for sale

```python
%%time
base_url = 'https://www.truecar.com/'
all_cars_urls = []
for link in pages:
    req = requests.get(link)

    soup = BeautifulSoup(req.text,'html.parser')

    req_links = soup.find_all('a')
    
    for link in req_links:
        suffix = str(link.get('href'))
        if ('used-cars-for-sale/listing/' in suffix):
            car_url = base_url + suffix
            all_cars_urls.append(car_url)
```

### Formatting the obtained links to avoid duplicates

```python
set_all_cars_urls = set(all_cars_urls)
n_urls = list(set_all_cars_urls)
len(n_urls)
```

### Adopting the columns for creating a dataframe

```python
make=[]
model=[]
trim=[]
year = []

mileage = []

location = []

engine=[]
transmission=[]
drive_type=[]
price = []
exterior_color=[]
interior_color=[]
vehicle_history=[]
MPG_city=[]
MPG_highway=[]
VIN=[]
accidents =[]
owners=[]
car_usage_type=[]
fuel_type=[]

i=0
for output in output_details:
    

    current_car = output.find('span',{'class':"vehicle-header-make-model text-truncate"}).get_text()
    
    try:
        make.append(current_car.split(' ')[0])
    except:
        make.append('n/a')
    
    try:
        model.append(current_car.split(' ')[1])
    except:
        model.append('n/a')
    
    try:
        trim.append(output.find('div',{'class':"font-size-1 text-truncate"}).get_text())
        
    except:
        trim.append('n/a')
    
    try:
        year.append(output.find('span',{'class':"vehicle-card-year font-size-1"}).get_text())
    except:
        year.append('n/a')
        
          
        
    # mileage
    try:
        mileage.append(output.find('div',{'data-test':"vehicleMileage"}).get_text())
    except:
        mileage.append('n/a')
    
        
    try:
        location.append(output.find('div',{'data-test':"vehicleCardLocation"}).get_text())
    except:
        location.append('n/a')    
        
    
    #price 
    try:
        price.append(output.find('div',{'data-test':"vehicleListingPriceAmount"}).get_text())
    except:
        price.append('n/a')
        
    clr=output.find('div',{'class':"vehicle-card-location font-size-1 margin-top-1 text-truncate"}).get_text()
    
    try:
        exterior_color.append(clr.split(',')[0].split(' ')[0])
    except:
        exterior_color.append('n/a')
    
    try:
        interior_color.append(clr.split(',')[1].split(' ')[1])
    except:
        interior_color.append('n/a')
    
    vhkl_hist=output.find('div',{'data-test':"vehicleCardCondition"}).get_text()
    
    try:
        accidents.append(vhkl_hist.split(',')[0])
    except:
        accidents.append('n/a')
        
    try:
        owners.append(vhkl_hist.split(',')[1])
    except:
        owners.append('n/a')
        
    try:
        car_usage_type.append(vhkl_hist.split(',')[2])
    except:
        car_usage_type.append('n/a')
        
        
    met = output.find('div',{'class':"font-size-1 margin-top-1"}).get_text()
    
    try:
        MPG_city.append(((met.split(',')[0]).split(':')[1]).split('-')[0])
    except:
        MPG_city.append('n/a')
        
    try:
        MPG_highway.append(((met.split(',')[0]).split(':')[1]).split('-')[1])
    except:
        MPG_highway.append('n/a')
        
    try:
        engine.append((met.split(',')[1]).split(':')[1])
    except:
        engine.append('n/a')
        
    try:
        transmission.append((met.split(',')[2]).split(':')[1])
    except:
        transmission.append('n/a')
        
    try:
        drive_type.append((met.split(',')[-1]))
    except:
        drive_type.append('n/a')
        
    try:
        ft=((met.split(',')[1]).split(':')[1]).split(' ')
        if ft[-1]=='Turbocharged' or ft[-1]=='Twin-Turbo':
            fuel_type.append(ft[-2])
        else: 
            fuel_type.append(ft[-1])
    except:
        fuel_type.append('n/a')
        
    try:
        VIN.append((output.find('div',{'class':"font-size-1 vehicle-card-vin-challenger"}).get_text()).split(':')[1])
    except:
        VIN.append('n/a')
```

### Obtaining the values of the dataset with results of one page list

```python

Used_Cars = pd.DataFrame({'Brand': make, 'Model': model, 'Trim' : trim, 'Manufature Year': year, 'Location':location, 
                          'Exterior Color' : exterior_color, 'Interior Color' : interior_color, 'Accidents':accidents,
                          'Owners':owners,'Car Usage Type':car_usage_type,'MPG City': MPG_city, 'MPG Highway':MPG_highway,
                          'Engine':engine, 'Transmission': transmission,'Drive type' : drive_type, 'Fuel Type': fuel_type, 'VIN':VIN, 'Mileage':mileage,'Price': price})
```
